```python
import pandas as pd
from imblearn.under_sampling import RandomUnderSampler
```

```python
df = pd.read_csv("./dataset.csv")
```

```python
df['target'].value_counts().plot.bar(title = "Dataset Distribution", ylabel="Counts")
```

```python
rus = RandomUnderSampler(random_state=123123)
```

```python
df['path'].to_numpy()
```

```python
X_resampled, y_resampled = rus.fit_resample(df['path'].to_numpy().reshape(-1, 1), df['target'].to_numpy().reshape(-1, 1))
```

```python
df = pd.DataFrame({'path': X_resampled.ravel(), 'target': y_resampled.ravel()})
```

```python
df['target'].value_counts().plot.bar(title = "Dataset Distribution", ylabel="Counts")
```

```python
df.to_csv("balanced_dataset.csv", index=False)
```
